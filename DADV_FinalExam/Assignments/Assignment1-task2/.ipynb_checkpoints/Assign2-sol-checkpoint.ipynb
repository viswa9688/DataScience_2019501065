```python
import pandas as pd
import numpy as np
import random
import glob
import csv, os
```

```python
def is_file_empty(file_path):
    return os.path.exists(file_path) and os.stat(file_path).st_size == 0
```

```python
count = 0
df = pd.DataFrame()
for files in glob.glob('C:\\Users\\HARIKA\\Desktop\\DataScience_2019501065\\DADV_FinalExam\\Assignments\\Assignment2\\*.csv'):
    if count == 30:
        break
    f=open(files, 'r')
    name = os.path.basename(f.name).replace(".csv","")
    if is_file_empty("C:\\Users\\HARIKA\\Desktop\\DataScience_2019501065\\DADV_FinalExam\\Assignments\\Assignment2\\"+name+".csv"):
        continue
    data = pd.read_csv(f)
    gain = (data.at[data.shape[0]-1, "Close Price"] / data.at[0,"Close Price"]) - 1
#     gain = (close[len(close)-1]/close[0])-1
    row = {'Code': name, 'Gain': gain}
    df = df.append(row, ignore_index=True)
    count = count + 1
    f.close()
print("done")   
```

```python
df = df.sort_values("Gain", ascending=False)
df.head(10)
```

```python
dic = {}
count = 1
for files in glob.glob("E:\\DADV\\BSE-Assignment\\Datasets\\*.csv"):
    f=open(files, 'r')
    if count == 10:
        break
    name = os.path.basename(f.name).replace(".csv","")
    if is_file_empty("E:\\DADV\\BSE-Assignment\\Datasets\\"+name+".csv"):
        continue
    data = pd.read_csv(f)
    dic[name] = list(data["Close Price"])
    count = count + 1
    f.close()

#making all equal sized lists (for correlation to happen)
dic["icici"] = list(icici["Close Price"])
max_length = 0
for lis in dic.values():
    max_length = max(max_length, len(lis))
for lis in dic.values():
    lis += [0.0] * (max_length - len(lis))

#finding the correlation
df = pd.DataFrame()
for each1 in dic.keys():
    corr = np.corrcoef(dic.get("icici"), dic.get(each1))[0][1]
    row = {'Code': each1, 'Corr': corr}
    df = df.append(row, ignore_index=True)
df
```

```python
df = df.sort_values("Corr", ascending=False)
df
```
